```python
#Package loading
import os
from binance import Client
import pandas as pd
import datetime
import json
import datetime as dt
```

## API setting

```python
api_key = "aaa"
api_secret = "aaa"
```

```python
#Client setting
client = Client(api_key, api_secret, tld='us')
```

## Data extraction

```python
#SET THE TOKEN:
symbol = "BTCUSDT"

#SET THE FREQUENCY OF THE DATA
interval="1d" #Every hour
# valid intervals - 1m, 3m, 5m, 15m, 30m, 1h, 2h, 4h, 6h, 8h, 12h, 1d, 3d, 1w, 1M

#Set the time period
ini= "1 Sept,2021"

fin = "1 Jun,2023" 
```

```python
#Extract the data
klines = client.get_historical_klines(symbol, interval, ini , fin)
```

```python
#Save the data into a dataframe
data = pd.DataFrame(klines)

# Add the colums name
data.columns = ['open_time','open', 'high', 'low', 'close', 'volume','close_time', 'qav','num_trades','taker_base_vol',"taker_quote_vol", 'ignore']

#Remove "ignore" because is deprecated
data = data[['open', 'high', 'low', 'close', 'volume','close_time']]
```

```python
data = data.copy()
```

```python
data.head()
```

```python
#Convert the timestamp
data["date"] = [dt.datetime.utcfromtimestamp(x/1000.0) for x in data.close_time]

#Remove the hours
data["date"]=data["date"].dt.strftime('%Y-%m-%d')
```

```python
data.head()
```

```python
#Select the needed columns
data = data[['open', 'high', 'low', 'close', 'volume', 'date']]
```

```python
#Final dataset
data
```

## Merge with market dominace

```python
#load the dataset
dominance= pd.read_csv("/content/dominance.csv")
```

```python
dominance.head()
```

```python
#Full outer join
Bitcoin = pd.merge(data, dominance, on = "date", how="outer")
```

```python
#Select the columns
Bitcoin = Bitcoin[['open', 'high', 'low', 'close', 'volume', 'date', 'dominance']]
```

```python
#Export the dataframe
Bitcoin.to_csv('Bitcoin_fin.csv', index = None, header=True)
```
